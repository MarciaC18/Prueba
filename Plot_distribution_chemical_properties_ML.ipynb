```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Cargar el archivo CSV
data = pd.read_csv('MDM2-P53-decoys_HBDA.csv')

# Función para calcular KDE normalizado
def plot_normalized_kde(data, column, xlabel, color):
    density = gaussian_kde(data[column], bw_method=0.5)
    xs = np.linspace(min(data[column]), max(data[column]), 1000)
    density_cdf = density(xs)
    density_normalized = density_cdf / max(density_cdf)
    
    plt.plot(xs, density_normalized, color=color)
    plt.fill_between(xs, density_normalized, color=color, alpha=0.5)
    plt.xlabel(xlabel)
    plt.ylabel("Densidad de probabilidad")
    plt.title(f'Distribución de {xlabel}')
    plt.ylim(0, 1)
    plt.grid(True)
    #plt.xticks(x_ticks_acc)
   

# Graficar la distribución de nHBAcc con densidad de probabilidad normalizada
plot_normalized_kde(data, 'nHBAcc', 'Nº de aceptores de H', 'orange')
plt.savefig('AceptorH_grafico.jpg')
plt.clf()  
# Graficar la distribución de nHBDon con densidad de probabilidad normalizada
plot_normalized_kde(data, 'nHBDon', 'Nº de donadores de H', 'orange')
plt.savefig('DonorH_grafico.jpg')

plt.show()

```
